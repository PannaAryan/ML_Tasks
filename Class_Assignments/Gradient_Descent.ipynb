Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Necessary Functions

```python
def cost_function(x):
    return x**2 - 4*x + 5

def gradient(x):
    return 2 * x - 4

def gradient_descent(initial_x, learning_rate, iterations):
    x_values = np.zeros(iterations)
    cost_values = np.zeros(iterations)
    x = initial_x

    for i in range(iterations):
        cost = cost_function(x)
        x_values[i] = x
        cost_values[i] = cost

        x -= learning_rate * gradient(x)

    return x_values, cost_values
```

Plot

```python
initial_x = 5.0
learning_rate = 0.1
iterations = 75

x_values, cost_values = gradient_descent(initial_x, learning_rate, iterations)

plt.plot(x_values, cost_values, marker='o', linestyle='-', color='red')
plt.title('Gradient Descent')
plt.xlabel('x values')
plt.ylabel(' Cost values')
plt.grid(True)
plt.show()
```
